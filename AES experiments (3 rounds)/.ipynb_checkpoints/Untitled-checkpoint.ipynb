```python
import random

# --- utility: dot product mod 2 of two 4-bit ints ---
def dot2(a: int, b: int) -> int:
    # parity of ones in (a & b)
    return bin(a & b).count("1") & 1

# --- test the integral property on a multiset M ---
def has_integral(M: list[int], beta: int, v: int) -> bool:
    total = sum(1 if dot2(beta, x)==0 else -1 for x in M)
    return (total % (2**v)) == 0


# --- build inverse S-box ---
def inverse_S(S: list[int]) -> list[int]:
    invS = [0]*16
    for x in range(16):
        invS[S[x]] = x
    return invS
```

```python
def find_linear_structures(Sbox):
    """
    Given Sbox: a list of 16 ints in [0..15],
    returns a list of (beta, delta) pairs where
      Sbox[x^beta] ^ Sbox[x] == delta  for all x.
    """
    invS = inverse_S(Sbox)
    n = 4
    N = 1 << n
    lin = {}
    for beta in range(1, N):
        found = []
        # collect all output‐differences for this beta
        for w in range(1,N):
            deltas = { dot2(beta, invS[x ^ w] ^ invS[x]) for x in range(N) }
            # if there's exactly one, it's constant => a linear structure
            if len(deltas) == 1:
                found.append(w)
        if found:
            lin[beta] = found
    return lin

# build your master list: one sub‐list per S‐box
all_structures = [
    find_linear_structures(Sbox)
    for Sbox in affine_equiv_classes_302
]

# example: print how many structures each S‐box has
for idx, struct in enumerate(all_structures):
    print(f"S-box #{idx:3d} has linear structures: {struct}")
```

```python

beta_list = list(range(1,16))   # all non-zero masks
v_list    = [2,3,4]
num_trials = 100
results = {}  # will map (beta,v,idx) -> list of valid keys

All_vector = list(range(16))
for beta in beta_list:
    print(beta)
    for v in v_list:
        for idx, Sbox in enumerate(affine_equiv_classes_302):
            #print(f"Sbox number {idx}")
            inv = inverse_S(Sbox)
            valid_keys = list(range(1, 16))
            # trial loop
            for _ in range(num_trials):
                # generate a valid M
                M = list(range(16))
                while not has_integral(M, beta, v) or M == All_vector:
                    M.append(random.randrange(16))
                if v == 3:
                    print(M)
                # test each key on a copy of valid_keys
                for k in valid_keys.copy():
                    M_prime = [inv[Sbox[x] ^ k] for x in M]
                    if not has_integral(M_prime, beta, v):
                        valid_keys.remove(k)
            
            results[(beta, v, idx)] = valid_keys
            #print(f"β={beta:04b}, v={v}, S-box#{idx} → preserving keys: {valid_keys}")

# print out results
# for (beta, v, sidx), keys in results.items():
#     print(f"β={beta:04b}, v={v}, S-box#{sidx} → preserving keys: {keys}")



```

```python
# Step 2: regroup by S-box index
results_by_sbox = {}
for (beta, v, sidx), keys in results.items():
    results_by_sbox.setdefault(sidx, {})[(beta, v)] = keys

# Step 3: print grouped by S-box
for sidx in sorted(results_by_sbox):
    print(f"\nS-box #{sidx:3d}:")
    bv_map = results_by_sbox[sidx]
    if not bv_map:
        print("  (no (β,v) pairs found)")
        continue
    for (beta, v), keys in sorted(bv_map.items()):
        print(f"  β = {beta:04b}, v = {v}  →  preserving keys: {keys}")
```

```python
# assume results_by_sbox is already defined as:
# { sidx : { (beta,v) : [keys...] , … } , … }

diffs = []  # will collect (sidx, beta, keys_v2, keys_v4)

for sidx, bv_map in results_by_sbox.items():
    for beta in beta_list:
        # get the two lists (or [] if missing)
        keys_v2 = bv_map.get((beta, 2), [])
        keys_v4 = bv_map.get((beta, 3), [])

        # compare as sets so ordering doesn't matter
        if set(keys_v2) != set(keys_v4):
            diffs.append((sidx, beta, keys_v2, keys_v4))

# report
if not diffs:
    print("For every S-box and β, the sets for v=2 and v=4 coincide.")
else:
    print("Differences found for (S-box, β):\n")
    for sidx, beta, k2, k4 in diffs:
        print(f" S-box #{sidx:3d}, β = {beta:04b}:")
        print(f"    v=2 → {k2}")
        print(f"    v=4 → {k4}\n")
```
